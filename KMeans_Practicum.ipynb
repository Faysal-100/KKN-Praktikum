# K-Means Clustering Practicum (Iris Dataset)
This notebook demonstrates K-Means clustering using a real dataset.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Actual_Species'] = iris.target
df.head()
```

```python
# Select features
X = df.drop(columns=['Actual_Species'])
```

```python
# Elbow Method
wcss = []
for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X)
    wcss.append(model.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
df.head()
```

```python
# Visualization
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['Cluster'])
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200)
plt.title('K-Means Clustering Result')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```
